### 모듈
* 변수, 함수, 클래스가 작성된 파이썬 파일(.py)

```python
# calculator.py

def plus(a, b):
    return a + b

def minus(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b
```

```python
# import
import module

module.변수1 # 호출
module.클래스1()
module.함수1()




# from ~ import
from module import 함수1

함수1() # 호출


import module.함수1 # error


# , *. 여러개를 불러올 때
from module import 클래스1, 함수1

from module import *




# as. import 하는 대상의 이름을 변형하여 사용할 수 있음
import module as mod

mod.변수1 # 호출
mod.클래스1()
mod.함수1()


from module import 함수1 as 함수

함수() # 호출
```

### 패키지
* 파이썬 파일(.py)들로 구성된 디렉토리 구조
* 패키지를 구성하는 모든 폴더는 __init__.py를 포함하는 것이 권장됨

```python
# stringLength.py
def strLen(str):
    return len(str)

# stringUpper.py
def strupper(str):
    return str.upper()

# stringLower.py
def strLower(str):
    return str.lower()
```

```python
# main.py
from package_sample import stringLength

print(stringLength.strLen('string'))
```

```python
from package_sample import *

print(stringLength.strLen('string'))
# NameError: name 'stringLength' is not defined
```

```python
# main.py
import package_sample

print(package_sample.stringLength.strLen('string'))
# AttributeError: module 'package_sample' has no attribute 'stringLength'
```
