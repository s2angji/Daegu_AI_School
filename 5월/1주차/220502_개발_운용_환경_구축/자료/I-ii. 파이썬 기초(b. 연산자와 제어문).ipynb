### 관계연산자(비교연산자)
* 둘의 값을 비교하기 위해 사용하는 연산자
* bool 형태의 값을 반환

|   표현식  | True를 반환하는 경우  |
|:-------:|:---|
|  x > y  | x가 y보다 큰 값일 때  |
|  x < y  | x가 y보다 작은 값일 때  |
|  x >= y | x가 y보다 크거나 같은 값일 때  |
|  x <= y | x가 y보다 작거나 같은 값일 때   |
|  x == y | x와 y가 같은 값일 때 (x is y 로 표현 가능)  |
| x != y | x와 y가 다를 때 (x is not y 로 표현 가능)   |

```python
x = 100
y = 300
print(type(x > y), x > y)
```

### 논리 연산자

```python
x = 15
condi_1 = x % 3 == 0
condi_2 = x % 5 != 0

print(condi_1, condi_2)
print('and', condi_1 & condi_2)
print('or', condi_1 | condi_2)
print('not', not condi_1)
```

### 반복문(for, while)
* 반복되는 작업을 처리할 수 있는 제어문
* 코드의 가독성을 좋게할 수 있음

```python
a = [1, 2, 3, 4, 5]
print(a[0])
print(a[1])
print(a[2])
print(a[3])
print(a[4])
```

```python
a = [1, 2, 3, 4, 5]
for i in a: # list a에 있던 요소(element) 값이 하나씩 i에 대입됨
    print(i)
```

* for문: 반복 횟수가 정해있는 경우 주로 사용
* while문: 반복 횟수를 알 수 없는 경우 주로 사용. 조건문으로 제어함

```python
# for
for 변수명 in iterable 자료형: 
    명령문

# while
while 조건문: # 조건문이 True인 동안 계속 명령문을 반복 수행함
    명령문
```

```python
# range(시작 숫자, 종료 숫자, step) # default 시작 숫자: 0, step: 1

# print(range(0, 10, 1))
for i in range(10):
    print(i)
```

```python
a = [(1, 10), (2, 20), (3, 30)]

for pair in a:
    print(pair)

print('-----')

# 쌍으로 이루어진 값을 분리해서 가져올 수 있음
for i, j in a:
    print(i, j)
    
print('-----')

# 몇번째 요소인지에 대한 순서를 함께 읽기 
for idx, pair in enumerate(a):
    print(idx, pair[0], pair[1])
```

```python
x = 0
while x < 10: # 10보다 작은 값만 출력하도록 조건문 설정
    print(x)
    x = x + 1 # 이 명령문이 없을 경우 무한 loop에 빠질 수 있어 주의해야 함
```

### 조건문(if)
* 상황 조건에 따라 다른 코드를 수행해야 할 때 사용

```python
# if
if 조건문:
    명령문
```

```python
# if ~ else
if 조건문:
    명령문1
else: # 조건문이 True가 아닐 때 실행
    명령문2
```

```python
# elif. 조건을 여러개 설정해야 할 때 사용
if 조건문1:
    명령문1
elif 조건문2:
    명령문2
    .
    .
    .    
else: # 모든 조건문이 False일 때 실행
    명령문n
```

* 조건문: bool 값(True, False)을 가지는/반환하는 식
* 조건문 위치에 있는 0, ''(빈문자)는 False, 그 외는 True로 취급됨

```python
x = 1
y = 10
if x > y:
    print('x가 y보다 크다')
else:
    print('x는 y보다 작다')
```

```python
a = ['apple', 'banana', 'candy', 'dolphin']
if 'apple' in a:
    print('apple exists in list')
```

##### continue와 break
* 반복문 수행을 제어하는 구문
* 주로 반복문 안에서 조건문과 함께 사용함   
      
      
* continue: 이후 명령문을 건너뛰고, 다음 반복문으로 넘어감
* break: 해당 반복문을 종료함

```python
for i in range(1, 11, 2):
    print(f'odd number: {i}')
```

```python
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(f'odd number: {i}')
```
