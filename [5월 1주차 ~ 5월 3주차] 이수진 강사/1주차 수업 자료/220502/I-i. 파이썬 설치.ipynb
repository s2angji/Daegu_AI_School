### Python이란?
* 플랫폼 독립, 인터프리터 언어
* 객체 지향, 동적 타이핑 언어
* 쉬운 프로그래밍
* AI, 통계, 데이터 분석을 비롯한 다양한 오픈 소스 라이브러리가 구현되어 있음

* 객체 지향적: 데이터(attribute)와 연산(method)으로 구성된 '객체'를 중심으로 프로그램 작성
* 동적 타이핑: 프로그램이 실행되면서, 프로그램 내 데이터 타입(자료형)을 검사

### Anaconda 설치하기
* 파이썬 패키지 관리 시스템
* 여러가지 과학 및 수학 라이브러리 포함하고 있으며 시각화에 유용함
* 가상환경을 편리하게 생성하고 관리할 수 있음       

다운로드 링크: www.anaconda.com/products/distribution

### PyCharm 설치하기
* 파이썬 언어에 특화된 프로그램 통합개발환경(IDE; Integrated development environment)
* 자동 완성, 자동 문법 검사, 줄맞춤, 실행, 디버깅 등 기능 제공      

다운로드 링크: www.jetbrains.com/ko-kr/pycharm/download

### Anaconda 가상환경 만들기
- Anaconda Prompt 사용       


가상환경 생성: conda create –n (가상환경 이름) python=(버전)   
가상환경 들어가기: conda activate (가상환경 이름)   

가상환경 나가기: conda deactivate   
가상환경 목록 확인: conda env list   
가상환경 삭제하기: conda env remove -n (가상환경 이름)   

라이브러리 설치: conda install (라이브러리 이름)   
라이브러리 설치(특정 버전): conda install (라이브러리 이름)==(버전)   
라이브러리 삭제: conda uninstall (라이브러리 이름)

#### 가상환경을 만드는 이유
* 어떤 프로젝트에 대해 필수적인 라이브러리를 확인하기 위해
* 프로젝트/라이브러리 간 버전 호환 문제를 방지하기 위해

### Jupyter notebook 사용하기
* Anaconda Prompt 사용
* Jupyter notebook 설치: conda install jupyter
* Anaconda 가상환경 연결 1: conda install ipykernel
* Anaconda 가상환경 연결 2: python -m ipykernel install --user --name (가상환경이름) --display-name (표시할 커널이름)
* Jupyter notebook 실행: jupyter notebook


* Anaconda 가상환경 연결 해지(가상환경 삭제x): jupyter kernelspec uninstall (커널이름)


### Jupyter notebook 단축키
* 위로 셀 추가: a
* 아래로 셀 추가: b
* 현재 셀 삭제: dd
* 삭제 취소: z
* 마크다운 셀로 변경: m
* 코드 셀로 변경: y     


* 아래셀과 병합: shift + m
* 현재 위치에서 셀 분할: ctrl + shift + -


* 현재 셀에 대한 코드 실행: ctrl + enter
* 현재 셀에 대한 코드 실행하면서 다음 셀로 이동: shift + enter
