```python
import sys

def list2dict(l):
    re_dict = {}
    
    for i in range(0, len(l), 2): # 0, 2, 4, 6 ...
        if l[i] == '':
            print(f'{int(i/2) + 1}번째 도서명 미기입')
            continue
        
        re_dict[l[i]] = int(l[i + 1])
    
    return re_dict

# 실행문
data_list = ['롤랑의 노래', 19800, '7년의 밤', '14500', '달 샤베트', 12000, 
             '', 13800, '백광', '14500']

print(list2dict(data_list))
# 수행 결과: 
# 4번째 도서명 미기입
# {'롤랑의 노래': 19800, '7년의 밤': 14500, '달 샤베트': 12000, '백광': 14500}
```

```python
import sys
import numpy as np

def list2dict(l):
    re_dict = {}
    
    l = np.array(l)
    l = np.reshape(l, (-1, 2))
    
    for idx, item in enumerate(l):
        if item[0] == '':
            print(f'{idx + 1}번째 도서명 미기입')
            continue
        
        re_dict[item[0]] = int(item[1])
    return re_dict

# 실행문
data_list = ['롤랑의 노래', 19800, '7년의 밤', '14500', '달 샤베트', 12000, 
             '', 13800, '백광', '14500']

print(list2dict(data_list))
# 수행 결과: 
# 4번째 도서명 미기입
# {'롤랑의 노래': 19800, '7년의 밤': 14500, '달 샤베트': 12000, '백광': 14500}
```

```python
import sys

class Converter:
    def __init__(self, value, unit):
        self.__value = value
        self.__unit = unit
        
    @property
    def value(self):
        return self.__value
    @value.setter
    def value(self, val):
        if val < 0:
            val = 0
        self.__value = val
        
    @property
    def unit(self):
        return self.__unit
    @unit.setter
    def unit(self, unit):
        if unit not in ['m', 'km']:
            unit = ''
        self.__unit = unit
    
        
    def cToM(self):
        if self.__unit != 'km':
            return
        self.__value = self.__value * 1000
        self.__unit = 'm'
    
    def cToKm(self):
        if self.__unit != 'm':
            return
        self.__value = round(self.__value / 1000, 3)
        self.__unit = 'km'
    
    def printValue(self):
        if self.__unit == '':
            return ''
        
        strV = list(str(self.__value))
        strV.reverse()

        retV = []

        if '.' in strV:
            split_idx = strV.index('.')
            retV.extend(strV[:split_idx + 1])
            strV = strV[split_idx + 1 : ]

        for idx, v in enumerate(strV):
            if idx > 0 and idx % 3 == 0:
                retV.append(',')
            retV.append(v)

        retV.reverse()
        return ''.join(retV) + ' ' + self.__unit

# 실행문
metric = Converter(1000, 'm')
print(metric.printValue()) # 1,000 m

metric.cToKm()
print(metric.printValue()) # 1.0 km

metric.value = 300
metric.unit = 'm'
print(metric.printValue()) # 300 m

metric.value = -50
print(metric.printValue()) # 0 m

metric.unit = 'cm'
print(metric.printValue()) # ''
```

### (5)
(4)에서 구현한 클래스를 상속하기

단위 mile를 추가 적용.   
(4)에서 선언한 함수가 변형되지 않도록 상속하여 사용   

(4)의 1), 2) 함수가 mile에 대해서도 제 기능을 하도록   
아래와 같이 변형함

1) m, mile을 km로 변환하는 함수, 변환된 값을 클래스내 저장   
예시)   
24000, 'm' -> 24, 'km'   
'm', 'mile'이 아닌 경우 변화 없음   
2) km, mile를 m로 변환하는 함수, 변환된 값을 클래스내 저장   
예시)   
2, 'km' -> 2000, 'm'   
'km', 'mile'이 아닌 경우 변화 없음   


아래 함수를 추가로 구현함   
5) m, km를 mile로 변환하는 함수. 변환된 값을 클래스내 저장   
예시)   
2, 'km' -> 1.243, 'mile'   
'm', 'km'가 아닌 경우 변화 없음   

* 1 mile = 1.609 km 로 계산(소수점 아래 4자리에서 반올림하여 사용)
* 4)의 setter 가능은 제거한 Converter 클래스 사용

```python
class Converter:
    def __init__(self, value, unit):
        self.value = value
        self.unit = unit
        
    def cToM(self):
        if self.unit != 'km':
            return
        self.value = self.value * 1000
        self.unit = 'm'
    
    def cToKm(self):
        if self.unit != 'm':
            return
        self.value = round(self.value / 1000, 3)
        self.unit = 'km'
    
    def printValue(self):
        if self.unit == '':
            return ''
        
        strV = list(str(self.value))
        strV.reverse()

        retV = []

        if '.' in strV:
            split_idx = strV.index('.')
            retV.extend(strV[:split_idx + 1])
            strV = strV[split_idx + 1 : ]

        for idx, v in enumerate(strV):
            if idx > 0 and idx % 3 == 0:
                retV.append(',')
            retV.append(v)

        retV.reverse()
        return ''.join(retV) + ' ' + self.unit
```

```python
class Converter_2(Converter):
    def __init__(self, value, unit):
        super().__init__(value, unit)
        
    def cToM(self):
        if self.unit == 'km':
            self.value = self.value * 1000
        elif self.unit == 'mile':
            self.value = self.value * 1.609 # mile to km
            self.value = self.value * 1000 # km to m
        
        self.value = round(self.value, 3)
        self.unit = 'm'
    
    def cToKm(self):
        if self.unit == 'm':
            self.value = self.value / 1000
        elif self.unit == 'mile':
            self.value = self.value * 1.609 # mile to km
        
        self.value = round(self.value, 3)
        self.unit = 'km'
    
    def cToMile(self): 
        if self.unit == 'm':
            self.value = self.value / 1000 # m to km
            self.value = self.value / 1.609 # km to mile
        elif self.unit == 'km':
            self.value = self.value / 1.609 # km to mile
        
        self.value = round(self.value, 3)
        self.unit = 'mile'
```

```python
# 실행문
metric = Converter_2(1000, 'm')
print(metric.printValue()) # 1,000 m

metric.cToMile()
print(metric.printValue()) # 0.622 mile

metric.value = 300
metric.cToM()
print(metric.printValue()) # 482,700.0 m

metric.cToKm()
print(metric.printValue()) # 482.7 km
```
